```python
# 补全中间港口在测试集trace的训练集数据（只考虑trace为2）

import pandas as pd
import numpy as np
import time
import warnings
warnings.filterwarnings('ignore')

start = time.process_time()
train_trace = ['CNSHK-GRPIR-ITSPE-ITGOA', 'HKHKG-FRFOS-FRMRS-TNTUN', 'HKHKG-FRFOS-TNTUN',
               'CNSHK-MYTPP-LKCMB-MVMLE', 'CNSHK-MYTPP-MUPLU', 'CNSHK-MYTPP-MUPLU-REREU', 'CNSHK-MYTPP-MUPLU-ZADUR', 'CNSHK-MYTPP-MVMLE',
               'CNSHK-MYTPP-MYPKG-TRMER', 'CNSHK-MYTPP-NGAPP', 'CNSHK-MYTPP-PGPOM', 'CNSHK-MYTPP-RELPT', 'CNSHK-MYTPP-REREU',
               'CNSHK-MYTPP-SOBBO', 'CNYTN-GRPIR', 'CNSHA-NGAPP', 'CNTAO-CNSHA-ESALG-DAKAR-KATI']

port_path = '../../../dataset/test_data/port.csv'
port_df = pd.read_csv(port_path)


# 获取分割时间
def obtain_port_time(df):
    df = df.loc[df['speed'] < 10].reset_index(drop = True)
    if (df.loc[0, 'TRANSPORT_TRACE'] == 'CNYTN-GRPIR'):
        df['TRANSPORT_TRACE'] = 'SGSIN-GRPIR'
    elif (df.loc[0, 'TRANSPORT_TRACE'] == 'CNSHA-NGAPP'):
        df['TRANSPORT_TRACE'] = 'CNSHA-SGSIN'
    elif (df.loc[0, 'TRANSPORT_TRACE'] == 'CNSHA-NGAPP'):
        df['TRANSPORT_TRACE'] = 'CNSHA-SGSIN'
    
    trace = df.loc[0, 'TRANSPORT_TRACE'].split('-')
    new_trace = trace[0] + '-' + trace[1]
    print('new_trace:{}'.format(new_trace))
    
    if (df.loc[0, 'TRANSPORT_TRACE'] == 'CNTAO-CNSHA-ESALG-DAKAR-KATI'):
        port = trace[0]
    else:
        port = trace[1]
        
    LonB, LatB = get_port_position(port)

    for i in range(len(df)):
        # print('data{}'.format(i))
        LonA = df.loc[i, 'longitude']
        LatA = df.loc[i, 'latitude']
        dis = distance(LonA, LatA, LonB, LatB)
        # print('dis{}:{}'.format(i, dis))
        if dis < 10:
            return df.loc[i, 'timestamp']

# 订单分割 
def process_order(df):
    train_df = pd.DataFrame()
    orders = df.loadingOrder.unique()
    for i in range(len(orders)):
        order_data = df.loc[df['loadingOrder'] == orders[i]].reset_index(drop = True)
        print('orders{}:{}'.format(i, order_data.loc[0, 'TRANSPORT_TRACE']))
        
        t = obtain_port_time(order_data)
        print('t：{}'.format(t))
        if (t != None):
            if (df.loc[0, 'TRANSPORT_TRACE'] == 'CNYTN-GRPIR'):
                order_data = order_data.loc[order_data['timestamp'] > t].reset_index(drop = True)
            else:
                order_data = order_data.loc[order_data['timestamp'] < t].reset_index(drop = True)
            
            order_data['loadingOrder'] = order_data.loc[0, 'loadingOrder'] + 'X'

            if (order_data.loc[0, 'TRANSPORT_TRACE'] == 'CNYTN-GRPIR'):
                order_data['TRANSPORT_TRACE'] = 'SGSIN-GRPIR'
            elif (order_data.loc[0, 'TRANSPORT_TRACE'] == 'CNSHA-NGAPP'):
                order_data['TRANSPORT_TRACE'] = 'CNSHA-SGSIN'
                
            trace = order_data.loc[0, 'TRANSPORT_TRACE'].split('-')
            new_trace = trace[0] + '-' + trace[1]
            order_data['TRANSPORT_TRACE'] = new_trace
        else:
            order_data = pd.DataFrame()
            
        train_df = train_df.append(order_data, ignore_index = True)

    return train_df

def df_split(train_data):
    df = train_data[train_data['TRANSPORT_TRACE'] == 'CNTAO-CNSHA-ESALG-DAKAR-KATI'].reset_index(drop = True)
    start_port = 'CNSHA'
    end_port = 'SGSIN'
    
    start_Lon, start_Lat = get_port_position(start_port)
    end_Lon, end_Lat = get_port_position(end_port)
    print('{}\t{}'.format(start_Lon, start_Lat))
    print('{}\t{}'.format(end_Lon, end_Lat))

    dis_start, dis_end = [], []
    for i in range(len(df)):
        LonA = df.loc[i, 'longitude']
        LatA = df.loc[i, 'latitude']
        dis = distance(LonA, LatA, start_Lon, start_Lat)
        dis_start.append(dis)
            
    for i in range(len(df)):
        LonA = df.loc[i, 'longitude']
        LatA = df.loc[i, 'latitude']
        dis = distance(LonA, LatA, end_Lon, end_Lat)
        dis_end.append(dis)
        
    index_start = dis_start.index(min(dis_start))
    index_end = dis_end.index(min(dis_end))
    print(min(dis_start))
    print(min(dis_end))
    
    df = df.iloc[index_start: index_end +1, :].reset_index(drop = True)
    df['TRANSPORT_TRACE'] = 'CNSHA-SGSIN'
    return df

# 经纬度转换为距离
def distance(LonA, LatA, LonB, LatB):
    EARTH_RADIUS = 6378.137  # 千米

    def rad(d):
        return d * np.pi/ 180.0

    radLatA = rad(LatA)
    radLatB = rad(LatB)

    a = radLatA-radLatB
    b = rad(LonA)-rad(LonB)

    s = 2 * np.arcsin(np.sqrt(np.power(np.sin(a / 2),2)+ np.cos(radLatA) * np.cos(radLatB)*np.power(np.sin(b / 2),2)))
    s = s* EARTH_RADIUS
    #  保留两位小数
    s = np.round(s * 100)/100
    return s

def get_port_position(port):
    for i in range(len(port_df)):
        if (port_df.loc[i, 'TRANS_NODE_NAME'] == port) or (port in port_df.loc[i, 'TRANS_NODE_NAME']):
            # print(port)
            port_lon = port_df.iloc[i, :]['LONGITUDE']
            port_lat = port_df.iloc[i, :]['LATITUDE']

            return port_lon, port_lat


train_data = pd.DataFrame()
for i in range(10):
    print('i:{}'.format(i))
    train_df = pd.read_csv('../../../dataset/train_data/train_wash{}.csv'.format(i), header=None)
    train_df.columns = ['loadingOrder', 'carrierName', 'timestamp', 'longitude',
                        'latitude', 'vesselMMSI', 'speed', 'direction', 'vesselNextport',
                        'vesselNextportETA', 'vesselStatus', 'vesselDatasource', 'TRANSPORT_TRACE']

    train_df = train_df.loc[train_df['TRANSPORT_TRACE'].isin (train_trace)]
    train_df = train_df[(train_df['direction'] <= 36000) & (train_df['speed'] < 60)]
        
    train_data = train_data.append(train_df, ignore_index=True)
    print('running file{}:{} seconds'.format(i, time.process_time() - start))
    
joint_df = df_split(train_data)
train_data = process_order(train_data)

train_data = train_data.append(joint_df, ignore_index = False).reset_index(drop = True)
train_data.to_csv('../../../dataset/train_data/train_wash11.csv', header = None, index=False)
print('finished')
```
