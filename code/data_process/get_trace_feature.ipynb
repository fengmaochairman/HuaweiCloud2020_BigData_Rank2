```python
# 获取路由统计特征
# 获取carrierName、起终点港口统计特征
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

def trace_compile(train_raw):
    test_raw = pd.read_csv('../../dataset/order_data/test_order_cluster.csv')

    train_trace = train_raw.TRANSPORT_TRACE.unique()
    train_trace_num = train_raw.trace.unique()
    train_trace.sort()

    test_trace = test_raw.TRANSPORT_TRACE.unique()
    test_trace_num = test_raw.trace.unique()
    test_trace.sort()
    print('train_trace:{}'.format(len(train_trace)))

    print('test_trace:{}'.format(len(test_trace)))


    x = []
    for i in range(len(test_trace)):
        if test_trace[i] not in train_trace:
            # x.append(test_trace[i])
            test_trace_df = test_raw.loc[test_raw['TRANSPORT_TRACE'] == test_trace[i]].reset_index(drop = True)
            trace_num = test_trace_df.loc[0, 'trace']
            if trace_num not in train_trace_num:
                x.append(test_trace[i])
    print('not_trace:{}'.format(len(x)))
    print(x)

    
def get_trace_feature(df):
    df['trace_Dir'] = df['Dir']
    df['trace_aver_speed'] =  df['aver_speed']
    df['trace_dis'] = df['dis']
    df['trace_start_anchor_time'] = df['start_anchor_time']
    df['trace_end_anchor_time'] = df['end_anchor_time']
    df['trace_middle_anchor_time'] = df['middle_anchor_time']
    df['trace_anchor_time'] = df['anchor_time']
    df['trace_label'] = df['label']
    
    df = df.drop(['Dir', 'aver_speed', 'dis', 'start_anchor_time', 'end_anchor_time', 'middle_anchor_time', 'anchor_time', 'label'], axis =1)
    
    tmp = df.groupby(['TRANSPORT_TRACE'], sort = False)
    agg_function = ['mean', 'max', 'min', 'median', 'std']
    agg_col = ['trace_Dir', 'trace_aver_speed', 'trace_dis', 'trace_start_anchor_time', 'trace_end_anchor_time',
               'trace_middle_anchor_time', 'trace_anchor_time', 'trace_label']

    group = tmp[agg_col].agg(agg_function).reset_index()
    group.columns = ['TRANSPORT_TRACE'] + ['{}_{}'.format(i, j) for i in agg_col for j in agg_function]
    group = group.fillna(0)
    trace_compile(df)
    
    df = df[['TRANSPORT_TRACE', 'start_port', 'end_port']]
    df = df.merge(group, on = 'TRANSPORT_TRACE', how = 'left')
    
    df = df.drop_duplicates(['TRANSPORT_TRACE'])
   
    df.to_csv('../../dataset/order_data/trace_feature.csv', index = False)
    return df

def get_carrierName_feature(df):
    df['carrierName_Dir'] = df['Dir']
    df['carrierName_aver_speed'] =  df['aver_speed']
    df['carrierName_dis'] = df['dis']
    df['carrierName_start_anchor_time'] = df['start_anchor_time']
    df['carrierName_end_anchor_time'] = df['end_anchor_time']
    df['carrierName_middle_anchor_time'] = df['middle_anchor_time']
    df['carrierName_anchor_time'] = df['anchor_time']
    df['carrierName_label'] = df['label']
    
    df = df.drop(['Dir', 'aver_speed', 'dis', 'start_anchor_time', 'end_anchor_time', 'middle_anchor_time', 'anchor_time', 'label'], axis =1)
    
    
    tmp = df.groupby(['carrierName'], sort = False)
    agg_function = ['mean', 'max', 'min', 'median', 'std']
    agg_col = ['carrierName_aver_speed', 'carrierName_dis', 'carrierName_start_anchor_time', 'carrierName_end_anchor_time',
               'carrierName_middle_anchor_time', 'carrierName_anchor_time', 'carrierName_label']

    group = tmp[agg_col].agg(agg_function).reset_index()
    group.columns = ['carrierName'] + ['{}_{}'.format(i, j) for i in agg_col for j in agg_function]
    group = group.fillna(0)

    group.to_csv('../../dataset/order_data/carrierName_feature.csv', index = False)
    return df

def get_vesselMMSI_feature(df):
    df['vesselMMSI_Dir'] = df['Dir']
    df['vesselMMSI_aver_speed'] =  df['aver_speed']
    df['vesselMMSI_dis'] = df['dis']
    df['vesselMMSI_start_anchor_time'] = df['start_anchor_time']
    df['vesselMMSI_end_anchor_time'] = df['end_anchor_time']
    df['vesselMMSI_middle_anchor_time'] = df['middle_anchor_time']
    df['vesselMMSI_anchor_time'] = df['anchor_time']
    df['vesselMMSI_label'] = df['label']
    
    df = df.drop(['Dir', 'aver_speed', 'dis', 'start_anchor_time', 'end_anchor_time', 'middle_anchor_time', 'anchor_time', 'label'], axis =1)
    
    tmp = df.groupby(['vesselMMSI'], sort = False)
    agg_function = ['mean', 'max', 'min', 'median', 'std']
    agg_col = ['vesselMMSI_aver_speed', 'vesselMMSI_dis', 'vesselMMSI_start_anchor_time', 'vesselMMSI_end_anchor_time', 
               'vesselMMSI_middle_anchor_time', 'vesselMMSI_anchor_time', 'vesselMMSI_label']

    group = tmp[agg_col].agg(agg_function).reset_index()
    group.columns = ['vesselMMSI'] + ['{}_{}'.format(i, j) for i in agg_col for j in agg_function]
    group = group.fillna(0)

    group.to_csv('../../dataset/order_data/vesselMMSI_feature.csv', index = False)
    return df

def get_stat_port_feature(df):
    df['start_port_aver_speed'] =  df['aver_speed']
    df['start_port_dis'] = df['dis']
    df['start_port_start_anchor_time'] = df['start_anchor_time']
    df['start_port_end_anchor_time'] = df['end_anchor_time']
    df['start_port_middle_anchor_time'] = df['middle_anchor_time']
    df['start_port_anchor_time'] = df['anchor_time']
    df['start_port_label'] = df['label']
    
    df = df.drop(['Dir', 'aver_speed', 'dis', 'start_anchor_time', 'end_anchor_time', 'middle_anchor_time', 'anchor_time', 'label'], axis =1)
    
    tmp = df.groupby(['start_port'], sort = False)
    agg_function = ['mean', 'max', 'min', 'median', 'std']
    agg_col = ['start_port_aver_speed', 'start_port_dis', 'start_port_start_anchor_time', 'start_port_end_anchor_time',
               'start_port_middle_anchor_time', 'start_port_anchor_time', 'start_port_label']

    group = tmp[agg_col].agg(agg_function).reset_index()
    group.columns = ['start_port'] + ['{}_{}'.format(i, j) for i in agg_col for j in agg_function]
    group = group.fillna(0)

    group.to_csv('../../dataset/order_data/start_port_feature.csv', index = False)
    return df

def get_end_port_feature(df):
    df['end_port_aver_speed'] =  df['aver_speed']
    df['end_port_dis'] = df['dis']
    df['end_port_start_anchor_time'] = df['start_anchor_time']
    df['end_port_end_anchor_time'] = df['end_anchor_time']
    df['end_port_middle_anchor_time'] = df['middle_anchor_time']
    df['end_port_anchor_time'] = df['anchor_time']
    df['end_port_label'] = df['label']
    
    df = df.drop(['Dir', 'aver_speed', 'dis', 'start_anchor_time', 'end_anchor_time', 'middle_anchor_time', 'anchor_time', 'label'], axis =1)
    
    tmp = df.groupby(['end_port'], sort = False)
    agg_function = ['mean', 'max', 'min', 'median', 'std']
    agg_col = ['end_port_aver_speed', 'end_port_dis', 'end_port_start_anchor_time', 'end_port_end_anchor_time',
               'end_port_middle_anchor_time', 'end_port_anchor_time', 'end_port_label']

    group = tmp[agg_col].agg(agg_function).reset_index()
    group.columns = ['end_port'] + ['{}_{}'.format(i, j) for i in agg_col for j in agg_function]
    group = group.fillna(0)

    group.to_csv('../../dataset/order_data/end_port_feature.csv', index = False)
    return df

def process_order_df(df):
    
    # 'CNNSA-MYTPP-SGSIN-ZACPT-CGPNR-GALBV-CMKBI'
    trace_df = df[df['TRANSPORT_TRACE'] == 'CNNSA-MYTPP-SGSIN-ZACPT-CGPNR-GALBV-CMKBI'].reset_index(drop = True)
    trace_num = trace_df.loc[0, 'trace']
    
    match_df = df[df['TRANSPORT_TRACE'] == 'CNNSA-CGPNR-CMKBI'].reset_index(drop = True)
    x = match_df[['start_anchor_time', 'end_anchor_time', 'middle_anchor_time', 'anchor_time', 'aver_speed', 'dis', 'label']].mean(axis=0)
    
    trace_df['start_anchor_time'] = x.start_anchor_time
    trace_df['end_anchor_time'] = x.end_anchor_time
    trace_df['middle_anchor_time'] = x.middle_anchor_time
    trace_df['anchor_time'] = x.anchor_time
    trace_df['aver_speed'] = x.aver_speed
    trace_df['dis'] = x.dis
    trace_df['label'] = x.label
    
    df = df[df['TRANSPORT_TRACE'] != 'CNNSA-MYTPP-SGSIN-ZACPT-CGPNR-GALBV-CMKBI'].reset_index(drop = True)
    df = df.append(trace_df, ignore_index = True).reset_index(drop = True)
    
    # CNYTN-CLSAI
    trace_df2 = df[df['loadingOrder'] == 'YL465830871055'].reset_index(drop = True)
    x2 = df[df['TRANSPORT_TRACE'] == 'CNYTN-CLSAI'].reset_index(drop = True)
    trace_df2['aver_speed'] = x2.loc[0, 'aver_speed']
    trace_df2['dis'] = x2.loc[0, 'dis']
    trace_df2['label'] = x2.loc[0, 'label']
    
    # CNYTN-HRRIJ
    trace_df3 = df[df['loadingOrder'] == 'EP333215739038'].reset_index(drop = True)
    x3 = df[df['loadingOrder'] == 'JG899236270837'].reset_index(drop = True)
    trace_df3['aver_speed'] = x3.loc[0, 'aver_speed']
    trace_df3['dis'] = x3.loc[0, 'dis']
    trace_df3['label'] = x3.loc[0, 'label']
    
    # CNYTN-VNVUT-SGSIN-FRLEH
    trace_df4 = df[df['loadingOrder'] == 'EA906723087784'].reset_index(drop = True)
    trace_df4['TRANSPORT_TRACE'] = 'CNYTN-VNVUT-SGSIN-FRLEH'
        
    y = pd.concat([trace_df2, trace_df3, trace_df4], axis = 0) 
    df = df[~df['loadingOrder'].isin (['YL465830871055', 'EP333215739038'])].reset_index(drop = True)
    df = df.append(y, ignore_index = True).reset_index(drop = True)
    
    print(df)
    return df


    
df = pd.read_csv('../../dataset/order_data/B_order_info.csv')
df = process_order_df(df)
df = df[df['aver_speed'] > 10].reset_index(drop = True)

# 计算统计特征
get_trace_feature(df)
get_carrierName_feature(df)
# get_vesselMMSI_feature(df)
get_stat_port_feature(df)
get_end_port_feature(df)
print('finished')
```
