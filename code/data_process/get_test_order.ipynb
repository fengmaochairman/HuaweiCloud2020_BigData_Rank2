```python
# 统计测试集各个订单信息
# 包括路由起终点聚类编号， 起终点港口经纬度坐标
import pandas as pd

def get_port_num(port):
    print(port)
    
    df = port_df[port_df['TRANS_NODE_NAME'] == port].reset_index(drop = True)
    return df.loc[0, 'cluster'], df.loc[0, 'LONGITUDE'], df.loc[0, 'LATITUDE']
        
def add_port_num(df):
    start_port_list, end_port_list, trace_list = [], [], []
    start_lon_list, start_lat_list, end_lon_list, end_lat_list = [], [], [], []
    for i in range(len(df)):
        trace = df.loc[i, 'TRANSPORT_TRACE'].split('-')
        start_port = trace[0]
        end_port = trace[-1]
        print('num{}:{}'.format(i, df.loc[i, 'TRANSPORT_TRACE']))

        start_port_num, start_lon, start_lat = get_port_num(start_port)
        end_port_num, end_lon, end_lat = get_port_num(end_port)
        trace = str(start_port_num) + '-' + str(end_port_num)
        print('start_port:{}\tend_port:{}'.format(start_port_num, end_port_num))

        start_port_list.append(start_port_num)
        end_port_list.append(end_port_num)
        trace_list.append(trace)

        start_lon_list.append(start_lon)
        start_lat_list.append(start_lat)
        end_lon_list.append(end_lon)
        end_lat_list.append(end_lat)

    df['start_port'] = start_port_list
    df['end_port'] = end_port_list
    df['start_lon'] = start_lon_list
    df['start_lat'] = start_lat_list
    df['end_lon'] = end_lon_list
    df['end_lat'] = end_lat_list
    df['trace'] = trace_list
    df = df.drop(['onboardDate', 'longitude', 'latitude', 'speed', 'direction'], axis = 1)
    print(df)
    
    return df
 
test_df = pd.read_csv('../../dataset/test_data/B_testData0807.csv')
test_df = test_df.dropna(how = 'any', axis = 0)
port_df = pd.read_csv('../../dataset/test_data/ports_cluster.csv')

test_null = test_df[test_df.isnull().values == True]
print('test_null:{}'.format(len(test_null)))

test_df = test_df.drop_duplicates(['loadingOrder']).reset_index(drop = True)
order_df = add_port_num(test_df)
order_df.to_csv('../../dataset/order_data/test_order_cluster.csv', index=False)
```
